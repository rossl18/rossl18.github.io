```python
import numpy as np
f = open('CH01PR27.txt')
x = []
y = []
for lines in f :
    a, b = lines.split()
    y.append(float(a))
    x.append(float(b))
```

```python
x_bar = sum(x)/len(x)
y_bar = sum(y)/len(y)
print(x_bar, y_bar)
```

```python
SSX = 0 #Calculate Regression Coefficients with 
SSXY = 0
for i in range(0, len(x)) :
    SSX = SSX + np.power((x[i] - x_bar), 2)
    SSXY = SSXY + (x[i] - x_bar)*(y[i] - y_bar)
print(SSX, SSXY)
b_one = SSXY/SSX
print(b_one)
b_zero = y_bar - b_one*x_bar
print(b_zero)
```

```python
import matplotlib.pyplot as plt #Plottting Regression Line and Data
x = np.array(x)
y = np.array(y)
x_g = np.linspace(35, 85, 500)
y_g = b_one*x_g + b_zero
plt.plot(x_g, y_g)
plt.xlabel("Age")
plt.ylabel("Muscle Mass")
plt.title("Age v. Muscle Mass Regression")
plt.scatter(x, y)
plt.show()
```
